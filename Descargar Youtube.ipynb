Descargar un video: Puedes descargar un video de YouTube utilizando el siguiente código:

```python
from pytube import YouTube

# URL del video que deseas descargar
video_url = "https://youtu.be/HXE169-n5pM?si=t7J0utHNxrzWMXTk"

# Crea un objeto YouTube con la URL
yt = YouTube(video_url)

# Descarga el video en la máxima resolución progresiva
yt.streams.get_highest_resolution().download()
```

Descargar un video de una lista de reproducción: Si deseas descargar todos los videos de una lista de reproducción, puedes hacerlo de la siguiente manera:

```python
from pytube import Playlist

# Get the playlist URL
playlist_url = "https://youtube.com/playlist?list=PLdxuOh58KNA6_o3VqvY-i2gUSfmptxzuT&si=EJQv7tCR8i20VmcP"

# Create a Playlist object
playlist = Playlist(playlist_url)

# Download all videos in the playlist
for video in playlist.videos:
    video.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first().download()
```
